# Propulsion and Rockets

<img src="imgs/rocket_conceptual.svg">

Let's start with some definition:

The word *propulsion* has as its root the Latin *propellere* - to drive before oneself.  Propulsion is used to *change an object's motion*.  Of course it is not the only way to change an objects motion - as Newton taught us in the 17th century any force upon an object will *accelerate* that object, changing its motion over time.

And this is a subtle but important point - an object's *motion*, or momentum, is what we are affecting with force, not its position.  This point - of the importance momentum or velocity and its evolution over time - will come up again and again in coming chapters.

There are many ways to propel and object, but we are going to focus on **rocket** and **rocket-like propulsion**.  In simplistic terms, a rocket is a **reaction engine** whos reaction mass (*propellant*) is all internal to the system.  So aircraft jet engines would not be in this category because they use air as a large part of their reaction mass.

Generally a rocket is more narrowly defined as a reaction engine that accelerates its propellant by gasdynamic means through a de Laval Convergent-Divergent nozzle.  We will spend most of our time there, but will slightly widen the scope to any device which creates thrust solely with its own propellant, even if it doesn't do that through traditional gasdynamic means.  Most **electric propulsion** falls into this "**rocket-like**" category.

The study of motion and its evolution in time is called *kinetics* and its central theorems are Newton's second and third laws.  While everyone knows these laws, I am going to spend just a moment on them because we will need a nuanced interpretation of them when we get to nozzle theory.

# Momentum, Thrust and Specific Impulse

## Newton's third law

Newton's third law simply states that

> For every action there is an equal and opposite reaction

Simply stated, if an object exerts a force on another, it feels a force itself that is equal to that force but in the opposite direction.  This is important for propulsion because it implies that one cannot change the motion of an object without imparting force on another object.  

### A brief interlude on the *inertialess* drive

It is important to remember Newton's third law anytime someone suggest an *inertialess* or *massless* propulsion system!  Some of these technologies (solar sails, electrodynamic tethers, laser propulsion) do in fact have solid physical basis.  

Electrodynamic tethers are actually exchanging momentum with (and thus exerting a force on) the Earth through interaction with its magnetic field.  

Solar sails and laser propulsion rely on exchanging momentum with photons.  How is this possible when photons have no mass?  It turns out that photons do have momentum...  Combing Einstein's $E=mc^2$ with the Planck relationship, $E = h\nu$, we see

$$p_{ph} = mc = \frac{E}{c} = \frac{h\nu}{c} = \frac{h}{\lambda}$$

So using photons as a momentum exchange medium is possible and doesn't violate conservation of momentum!

# Newton's second law
The popular form of Newton's second law states simply $F = ma$.  And while revolutionary in the 17th century, it now seems an almost trivial piece of our technical knowledge as engineers.  But it is more subtle than it appears.

The differential form of the second law is $\vec{F} = m\frac{d\vec{V}}{dt}$ and in that form we already begin to see the central concept of changing motion (velocity) in time.  In reality this form is a special case of the more general $\vec{F} = \frac{d\vec{p}}{dt}$ where $\vec{p}$ is momentum.  

The kinematic assumption is that all dynamic systems consist of particles of constant mass in time ($\frac{dm}{dt} = 0$) and of course it is trivial to see that this general form reduces to the special (kinematic) form above under this constraint.  I only make this digression because the more general form of Newton's second law will become very useful when relax the kinematic constraint to look at nozzle theory. 

# Conservation of momentum

Looking at two particles and combinging Newton's first and second laws allows us to derive conservation of momentum:

$$F_1 = -F_2 = d\vec{p}_1 = - d\vec{p}_2$$

or

$$d\vec{p}_1 = -d\vec{p}_2$$

This is true of two particles interacting.  For a closed system with no external forces one can state that the total momentum of the system doesn't change or:

$$\sum_{N}d\vec{p} = 0$$

In rockets, it will often be more useful to use the control volume form of conservation of momentum which allows for force at system boundaries and mass moving in or out. Given a control volume, $CV$, with external control surface, $CS$, we can write the *integral form*:

$$\sum\mathbf{F}_{CV} = \frac{d}{dt} \int_{CV}u \rho dV + \oint_{CS} \rho u ( \mathbf{U} \cdot \mathbf{n} )dA$$

$\mathbf{n}$ is the outward-facing normal of the control surface.

Note that we have switched $U$ for velocity so as not to confuse it with $V$ or $v$ (volume or specific volume).  Also we have switched to bold-face notation for vectors to keep equations cleaner.  We will continue this notation from here on.

One can read this as:

> The sum of forces acting on the control volume is equal to accumulation of momentum in the volume due to unstead flow (think total mass of control volume changing in time) plus the rate at which momentum is being carried across the control surface surrounding the control volume.
    
The unsteady term in the conservation of momentum equation is useful for dynamic compressible flow problems (like acoustics or vortex shedding) or when the control volume itself is moving (for instance if you are tracking a moving piston or pump impeller).  For most of the propulsion problems we will be looking at, we can treat flow as steady and assume the unstead term is zero.

# Rocket Thrust

Let's make some thrust!

<img src='imgs/nozzle_pressure.svg'>

The figure above (blatently stolen from Prof. Cantwell's lecture notes) shows rocket chamber at rest on a frictionless sliding platform.  There is a control volume drawn around the chamber.  The question is, how much force (thrust) does the rocket produce?

### Conservation of momentum

Starting with our unsteady conservation of momentum equation:

$$\sum\mathbf{F}_{CV} = \frac{d}{dt} \int_{CV}u \rho dV + \oint_{CS} \rho u ( \mathbf{U} \cdot \mathbf{n} )dA$$

we can immediately start simplifying some things.  The first thing we will do is recognize that we are only interested in $x$ axis.  So we can look exclusively at forces, velocities and pressure in the $+/-x$ direction.  Since gravity acts vertically, we can ignore it as an external force when looking at acceleration in the $x$ direction.

We will also make the assumption of *steady flow* which means that there is not an accumulation / depletion of mass within the control volume over time.  This allows us to ignore the left integral above (accumulation term) as we discussed earlier:

$$\sum\mathbf{F}_{CV} = \oint_{A_c} \rho u ( \mathbf{U} \cdot \mathbf{n} )dA_c$$

Now if we look at the control surface inside the chamber, we can see that the internal pressure forces are balanced on either side of the control surface everywhere except the nozzle exit. There we have $P_0$ exerting force on the CS in the -x direction and $P_e$ in the +x direction, both over the area $A_e$.  Or

$$\int_{A_e} \overline{P_0} \cdot \mathbf{n}dA = -P_0 A_e$$

and

$$\int_{A_e} \overline{P_e} \cdot \mathbf{n}dA = P_e A_e$$

Since the rocket is reacting T in the -X direction:

$$\sum\mathbf{F}_{CS} = -T - P_0 A_e  + P_e A_e$$

The surface integral of momentum in the equation above is the *momentum flux* across the control surface around the chamber.  Since we know area-averaged velocity and density, $\overline{U_e}$ and $\overline{\rho_e}$ through the chamber exit area, the integral simplifies to an algebraic relation.  Note that $u_x = \overline{U}_e$ and that $\mathbf{U} \cdot \mathbf{n} = -\overline{U}_e$ since the exit area normal, $\mathbf{n}$ is in the -x direction and $\overline{U_e}$ is in the +x direction.  Substituting in and solving the integral gives us:

$$-T + P_e A_e - P_0 A_e = -\overline{\rho_e} \overline{U}_e^2 A_e$$

$$T = \overline{\rho_e} \overline{U}_e^2 A_e + (P_e - P_0) A_e$$

### The rocket thrust equation

Out of convenience, let us define

$$\dot{m} = \overline{\rho_e} \overline{U}_e A_e$$

Now we can write **the rocket thrust equation** in its typical form

### $$T = \dot{m}\overline{U}_e + (P_e - P_0) A_e$$

Some observations:
* Thrust depends on the *external pressure* that a rocket operates in and the same rocket produces more thrust at lower external pressure
* $\overline{\rho_e}$, $\overline{U}_e$ and $P_e$ are all related in a complex way by the *isentropic flow relations* which we will discuss more soon


## Exhaust Velocity, Effective Exhaust Velocity and Specific Impulse

Exhaust velocity is simply $U_e$ above.  It seems clear somehow that higher exhaust velocites are desireable as it enables us to produce **more thrust per unit propellant mass flow**.  However the pressure term complicates using that as a metric alone.  So instead we will define a new parameter we call *effective exhaust velocity*, $C$:

$$C = \frac{T}{\dot{m}}$$

$C$ is now a very useful parameter for us to compare the relative performance of rockets because it tells us directly how much thrust we can produce with a given *mass flowrate* of propellant.  When we get to the rocket equation, we will see how important mass is in rockets.

We will also define a closely related parameter that is more often used in describing rocket performance, *specific impulse*:

$$I_{sp} = \frac{C}{g_0}$$

Where $g_0$ is the standard gravitational parameter and the units of $I_{sp}$ are in seconds.  

$I_{sp}$ is sort of strange parameter at first glance but it is intuitively useful because it immediately tells us:

* The ratio of force to weight-flowrate for a rocket.  Given $I_{sp} = 200s$ one could say "The rocket produces 100 lbf of thrust for every lbm/s of propellant pushed through it".  --or--
* Given $I_{sp} = 200s$, if I have 1 lbm. of propellant and a one lbf. thruster, it can produce that thrust for 200 seconds of burn time

Suffice to say these lbm and lbf units are old and kind of shitty.  But you will see this all over in the propulsion literature.  Even people who stay metric with $N$ and $kg$ will regulary use $I_{sp}$ in units of $\frac{N}{kg-g_0}$ or $s$.

So now we can finally say:

### $$C = I_{sp} g_0 = \overline{U}_e + (P_e - P_0) \frac{A_e}{\dot{m}}$$

# Optimal nozzle expansion

<img src="imgs/nozzle_pressure.png">

The **area ratio** or **expansion ratio** ($\epsilon$) of a nozzle is the nozzle exit area, $A_e$ over its throat area, $A_t$ or $A^*$ where \* indicates the sonic condition.

$$\epsilon = \frac{A_e}{At} = \frac{A_e}{A^*}$$

By changing $\epsilon$, we can trade exit pressure, $P_e$ for exit velocity, $U_e$.  The question is, how should we chose $\epsilon$ to maximize effective exhaust velocity, $C$, and therefore thrust for our rocket?

We can re-arrange the equation for effective exhaust velocity above and substitute for $\dot{m}$ to get

$$C = \overline{U}_e \left[1 + \frac{P_e A_e}{\rho_e \overline{U}_e^2 A_e}\left(1-\frac{P_0}{P_e}\right)\right]$$

To simplify further, we will need to jump a bit ahead and use a result from the so-called *isentropic flow relations*.  We will look at these in more detail later, but we can use the following results to help:

$$\frac{P_e}{\rho_e U_e^2} = \frac{1}{\gamma M_e^2}$$

and

$$U_e^2 = \frac{\gamma R_u T_0 M_e^2}{1+\frac{\gamma-1}{2}M_e^2}$$

where $\gamma$ is the ratio of specific heats:

$$\gamma = \frac{C_p}{C_v}$$

For most high temperature rocket exhaust gasses $\gamma$ ranges between 1.1 and 1.3.  We will use 1.2 here.

This gives us 

$$C = \sqrt{\frac{\gamma R T_0 M_e^2}{1+\frac{\gamma-1}{2}M_e^2}} \left[1 +\frac{1}{\gamma M_e^2}\left(1-\frac{P_0}{P_e}\right)\right]$$

where $T_0$ is the gas stagnation (or **total**) temperature.  We'll see this equation again, but it's worth pointing out that the two dominant physical quantities that effective exhaust velocity, and hence $I_sp$ depend on are the molecular weight of the gas ($R = R_u / M_w$) and the stagnation temperature, $T_t$.  It also depends on $\gamma$ but to a lesser extent.

And one final isentropic flow to close it out:

$$M_e^2 = 2\left(\frac{\frac{Pe}{Pt}^{\frac{1-\gamma}{\gamma}}-1}{\gamma-1}\right) = 2\left(\frac{\left(\frac{P_e}{P_0}\frac{P_0}{P_t}\right)^{\frac{1-\gamma}{\gamma}}-1}{\gamma-1}\right)$$

Combining all of this together, we can look at what $C(\frac{P_e}{P_0})$ looks like:

```python
Pt_P0v = np.linspace(25, 100, 4)
Pe_P0v = np.linspace(0.1, 3, 100)
gamma = 1.2
MW = 15              #g / mol
R = 8.314e3 / MW
Tt = 2500

plt.figure(figsize=(10,5))
for Pt_P0 in Pt_P0v:
    Me2 = 2. * ( (Pe_P0v / Pt_P0) ** ((1. - gamma) / gamma) - 1.) / (gamma - 1)
    Ue = np.sqrt(gamma * R * Tt * Me2 / (1. + (gamma - 1.) / 2. * Me2))
    C = Ue * (1. + 1. / gamma / Me2 * (1. - 1. / Pe_P0v))
    plt.plot(Pe_P0v, C, label='Pt/P0 = %.1f' % Pt_P0)

plt.plot([1.0, 1.0], [2400, 3000], 'm--')
plt.legend(loc='best')
plt.xlabel('$P_e/P_0$')
plt.ylabel('C')
plt.ylim(2400,3000)
```

Note that $C$, and thus thrust, is maximized when $\frac{P_e}{P_0} = 1$ or $P_e = P_0$.

This says that, in general, we want to design a rocket nozzle to match the exit pressure to ambient pressure.  This is difficult to do for a rocket ascending through the atmosphere where the pressure is continuosly changing.  For traditional nozzles, a compromise that looks at the average performance over the ascent pressure profile is often chosen.
