# Chemical propulsion types

Chemical propulsion relies on the heat liberted by one or more propellants upon in an exothermic chemical reaction.  Usually two (or more) reactive chemicals are combined together in some form of **RedOx** reaction.  In the case of some monopropellant and solid propellants, the propellant may consist of a single, unstable molecule that decomposes exothermically.

The four primary types of chemical propulsion are:

* __Solid Rocket__ - Propellants combined into a "grain" in the solid phase that burn ballistically
* __Liquid Bipropellant Rocket__ - A liquid fuel and oxidizer are introduced in a combustion chamber where they react
* __Liquid monopropellant Rocket__ - A single liquid containing both oxidizing and reducing elements (either in solution or in a single molecule) decomposes exothermically.  Generally the reaction is initiated and sustained with a catalyst
* __Hybrid Rocket__ - A liquid phase propellant is introduced to a solid phase propellant within the combustion chamber where they react

## Chemical propulsion performance

We're going to take a brief interlude to discuss two useful parameters in characterizing rocket performance, $c^*$ and $c_f$.

$c^*$ is a convenient parameter because it only depends on chamber conditions (no dependency on nozzle geometry) and is directly measureable in test.  It is almost entirely a function of the propellant chemistry, although there is a weak depence on chamber pressure $P_c$.  It is defined as:

$$c^* = \frac{P_c A_t}{\dot{m}}$$

where $P_c$ is chamber total (stagnation) pressure, $A_t$ is the nozzle throat area and $\dot{m}$ is the total mass flowrate of propellants.  Its units are velocity.  For an ideal gas, $c^*$ can be computed simply from gas properties and stagnation temperatue:

$$c^* = \sqrt{\frac{RuT_0}{\gamma M_{w} \left(\frac{2}{\gamma+1}\right)^{\frac{\gamma+1}{\gamma-1}}}}$$

Using detailed chemical equillibrium codes such as CEA (http://www.grc.nasa.gov/WWW/CEAWeb/) or Propep (http://rocketworkbench.sourceforge.net/equil.phtml), $c^*$ can be computed for any propellant combination.  So-called **$c^*$ efficiency** is the primary metric by which chemical rockets combustion efficiency is measured.

$c^*$ is also a very useful design parameter.  Once it is known for a combination of propellants, it ties chamber pressure, nozzle throat area and propellant mass flowrate together in the design process.

$c_f$ is the **nozzle coefficient** or **thrust coefficient**.  It is related to $c^*$ and $C$:

$$C = c_f c^* $$

And depends on nozzle geometry and $\gamma$ for an ideal non-reacting flow.

There is a closed form solution for $c_f$ for ideal, non-reacting flows but we won't go into it here.  Suffice to say that:

$$c_f = f\left(\gamma, \epsilon, \frac{P_e}{P_0}\right)$$

### Rocket performance testing

It is often difficult to replicate perfectly mission atmospheric conditions that a rocket will see in ground testing.  It is also often difficult to measure thrust directly in a rocket test for various reasons.

It also turns out that flow in a nozzle is **relatively** easy to model while modeling combusting flow in a rocket chamber is extremely difficult, even today.  So often times rocket combustion chambers are tested without an optimized nozzle and then their performance in flight is extrapolated based on analytical or semi-analytical nozzle models.

Thus the utility of separating $C$ into $c^*$ and $c_f$ - we can measure and optimize $c^*$ easily and accurately in a combustion chamber ground test and model $c_f$ to understand what our in-flight performance will be.

### Comparing propellants

Going back to our equation for effective exhaust velocity

$$C = \sqrt{\frac{\gamma R T_0 M_e^2}{1+\frac{\gamma-1}{2}M_e^2}} \left[1 +\frac{1}{\gamma M_e^2}\left(1-\frac{P_0}{P_e}\right)\right]$$

we make the following observations / assumptions:

* $M_e$ is primarily a function of the nozzle expansion ratio and $\gamma$ (this is a result of isentropic flow)
* $\frac{P_0}{P_e}$ is, again, primarily a function of the nozzle (and chamber pressure)
* $\gamma$ generally varies between 1.1 and 1.4 for most species and $C$ is not terribly sensitive to it

With these assumptions, we will define $C_{max}$ where $\frac{P_0}{P_e}=1$ as:

$$C_{max} = \sqrt{\frac{\gamma R T_0 M_e^2}{1+\frac{\gamma-1}{2}M_e^2}} $$

And by our other assumptions we can say that:

$$C_{max} \sim \sqrt{\frac{T_0}{M_w}}$$

This is a hugely important result because it tells us that to maximize effective exhaust velocity or specific impulse, we want a propellant combination that
* Generates very high combustion temperature, $T_0$
* Has low molecular weight $M_w$

High combustion temperatures mean large $\Delta H^o_{RXN}$.  To first order $\Delta H^o_{RXN}$ depends on the relative bond strengths of reactants and products in the chemical reaction.  And for the most part, bond strengths are related to difference in electronegativity of reaction constituents.  Below is the periodic table with electronegativity called out.

To maximize $C$, we want propellants with low molecular mass (high on periodic chart) with large difference in electronegativities.  

That's why most chemical propellants include an oxidizing species containing oxygen, chlorine, and sometimes even Flourine.  And a reducing species that includes lost of hydrogen and sometimes low electronegativity metals such as Lithium, Aluminum or even Berylium.
